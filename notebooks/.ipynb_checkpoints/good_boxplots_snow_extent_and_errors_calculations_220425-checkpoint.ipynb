## Mes fichiers d'extension mensuelle moyenne de neige pour tous les modèles - Fichiers historiques

```python
# méthode avec fichiers historiques
histo = pd.read_csv('~/private-storage/fichiers_csv/Comparaison entre modèles - avec_glaciers_historical(3).csv', sep=',')
histo
```

```python
# passage au format long
histo_long = histo.melt(id_vars='model', var_name='Month', value_name='snow_extent')
histo_long
```

```python
ref_data = histo_long[histo_long['model'] == 'Reference']
ref_data
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x='Month', y='snow_extent', data=histo_long, color='lightblue')
sns.scatterplot(x='Month', y='snow_extent', data=ref_data, color='red', s=50, label='Reference', zorder=10)
plt.title('Historical', fontsize='30')
plt.ylabel('Snow Cover Extent (km2)', fontsize='25')
plt.xlabel('Month', fontsize='25')
plt.xticks(rotation=45, fontsize='20')
plt.yticks(fontsize='20')
plt.legend(fontsize='18')
plt.tight_layout()
plt.savefig("figures/boxplot_historical_snow_extent.png")
plt.show()
```

```python
outliers = []

# detection des outliers pour chaque mois
for mois in histo_long['Month'].unique():
    valeurs = histo_long[histo_long['Month'] == mois]['snow_extent']
    q1 = valeurs.quantile(0.25)
    q3 = valeurs.quantile(0.75)
    iqr = q3 - q1 #(interquantile range)
    seuil_bas = q1 - 1.5 * iqr
    seuil_haut = q3 + 1.5 * iqr

    # selection des outliers
    outliers_mois = histo_long[
        (histo_long['Month'] == mois) &
        ((histo_long['snow_extent'] < seuil_bas) | (histo_long['snow_extent'] > seuil_haut))
    ]
    outliers.append(outliers_mois)

outliers_df = pd.concat(outliers)
```

```python
plt.figure(figsize=(10, 6))

#boxplot des valeurs de neige
sns.boxplot(x='Month', y='snow_extent', data=histo_long, color='lightblue')

#mise en évidence des valeurs de référence
sns.scatterplot(x='Month', y='snow_extent', data=ref_data, color='red', s=50, label='Reference', zorder=10)

#mise en évidence des outliers
sns.scatterplot(x='Month', y='snow_extent', data=outliers_df, color='black', s=50)

#ajout des noms des modèles
for i, row in outliers_df.iterrows():
    plt.text(x=row['Month'], y=row['snow_extent'] + 1, s=row['model'], horizontalalignment='center',
             fontsize=8, color='darkred')

plt.title('Snow Extent')
plt.ylabel('snow extent (km2)')
plt.xticks(rotation=45)
plt.show()
```

```python
outliers_df
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Month', y='snow_extent', data=histo_long, color='lightblue')
sns.scatterplot(x='Month', y='snow_extent', data=ref_data, color='red', s=50, label='Reference', zorder=10)
plt.title('Snow Extent')
plt.ylabel('snow extent (km2)')
plt.xticks(rotation=45)
plt.show()
```

## Attribution d'un score à chaque modèle

### Exponentielle décroissante

```python
min_val = classement_modeles.min()
max_val = classement_modeles.max()
```

```python
scores_exp = np.exp(-(classement_modeles - min_val)/(max_val - min_val))
scores_exp
```

### Gaussienne "arrangée"

```python
mu = classement_modeles.min()
sigma = (classement_modeles.max() - classement_modeles.min())/4
```

```python
scores_gauss = np.exp(-((classement_modeles - mu) ** 2)/(2 * sigma **2))
```

```python
scores_gauss
```

```python
import numpy as np
from scipy.interpolate import make_interp_spline

# Indices numériques des modèles triés
x = np.arange(len(scores_gauss))
y = scores_gauss.values

# Interpolation spline pour lisser
x_new = np.linspace(x.min(), x.max(), 300)
spl = make_interp_spline(x, y, k=3)
y_smooth = spl(x_new)

# Tracé lissé
plt.figure(figsize=(12, 6))
plt.plot(x_new, y_smooth, color='coral')
plt.scatter(x, y, color='blue')  # Optionnel : points originaux
plt.xticks(x, scores_gauss.index, rotation=90)
plt.ylabel('Score Gaussien')
plt.title('Scores Gaussiens (courbe lissée)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
mu = classement_modeles.min()
sigma = (classement_modeles.max() - classement_modeles.min()) / 4

# Axe x pour la courbe théorique
x = np.linspace(classement_modeles.min(), classement_modeles.max(), 500)
gauss_curve = np.exp(-((x - mu) ** 2) / (2 * sigma ** 2))

# Scores déjà calculés
scores_gauss_values = scores_gauss.values
classements = classement_modeles.values

# Tracé
plt.figure(figsize=(12, 6))
plt.plot(x, gauss_curve, label='Score théorique (gaussienne)', color='purple')
plt.scatter(classements, scores_gauss_values, color='orange', label='Scores calculés', zorder=5)
plt.axvline(mu, linestyle='--', color='gray', label='μ (meilleur rang)')
plt.title("Fonction de Score Gaussienne et Scores des Modèles")
plt.xlabel("Classement modèle")
plt.ylabel("Score")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
mu = classement_modeles.min()  # Meilleure erreur
sigma = (classement_modeles.max() - classement_modeles.min()) / 4

# Score calculé
scores_gauss = np.exp(-((classement_modeles - mu) ** 2) / (2 * sigma ** 2))

# Générer la courbe théorique
x = np.linspace(classement_modeles.min(), classement_modeles.max(), 500)
gauss_curve = np.exp(-((x - mu) ** 2) / (2 * sigma ** 2))

# Tracé
plt.figure(figsize=(12, 6))
plt.plot(x, gauss_curve, label='Theoretical scoring function', color='blue')
#plt.scatter(classement_modeles.values, scores_gauss.values, color='red', zorder=5, label='Models scores')
plt.axvline(mu, color='gray', linestyle='--', label='μ (best error)')
plt.xlabel("Model error")
plt.ylabel("Gaussian score")
#plt.title("Transformation Gaussienne des Erreurs des Modèles")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
#plt.savefig("courbe_gaussienne.png")
plt.show()
```

```python
erreurs = scores_modeles["Erreur_moyenne"]

# Inverse simple
scores_inv = 1 / erreurs

# Inverse carré
scores_inv2 = 1 / (erreurs ** 2)

# Min-max scaling
scores_minmax = 1 - (erreurs - erreurs.min()) / (erreurs.max() - erreurs.min())

# Softmax inversé
scores_softmax = np.exp(-erreurs) / np.exp(-erreurs).sum()

scores_modeles["Scores_inv"] = scores_inv
scores_modeles["Scores_inv2"] = scores_inv2
scores_modeles["Scores_minmax"] = scores_minmax
scores_modeles["Scores_softmax"] = scores_softmax
```
